```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('cs_papers.csv')

# Create a new column that calculates the word count of each abstract
df['word_count'] = df['abstract'].apply(lambda x: len(str(x).split()))

# Plot: Distribution of Abstract Word Counts
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='word_count', bins=30, kde=True, color='purple')
plt.title('Distribution of Abstract Word Counts in Dataset')
plt.xlabel('Number of Words in Abstract')
plt.ylabel('Frequency (Number of Papers)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

 
print(f"Average abstract length: {df['word_count'].mean():.0f} words")
print(f"Longest abstract: {df['word_count'].max()} words")
```
