```python
import google.generativeai as genai
from IPython.display import display, Markdown, Image as IPImage
from PIL import Image, ImageDraw
import time

# 1. API Configuration
API_KEY = "MY_API_KEY"
genai.configure(api_key=API_KEY)
model = genai.GenerativeModel("gemini-2.0-flash-lite")

# 2. Generate a Local Image Programmatically 
image_path = "Sample.jpg"

# Create a 300x300 pixel image with a green background
img = Image.new("RGB", (300, 300), color="green")

# Draw a yellow circle right in the center
draw = ImageDraw.Draw(img)
draw.ellipse([100, 100, 200, 200], fill="yellow")

# Save it in local notebook directory
img.save(image_path)

# 3. Load and Display the created Image
display(IPImage(filename=image_path))

# 4. Multimodal Prompting
user_prompt = "Identify the geometric shape, its color, and the background color in this image."
print(f"User: {user_prompt}\n")

start_time = time.time()
response = model.generate_content([user_prompt, img])
latency = time.time() - start_time

# 5. Display Output and Metrics
display(Markdown(f"**Gemini:** {response.text}"))
print(f"\n[System Metric] Response Time: {latency:.2f} seconds")
```
